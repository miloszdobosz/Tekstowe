### Zadanie 1

```python
def delta(char1, char2):
    return 0 if char1 == char2 else 1
    
```

```python

def alghorithm(word1, word2, delta=delta):
    n1, n2 = len(word1) + 1, len(word2) + 1

    dist = np.empty((n1, n2), dtype=int)
    op = np.zeros_like(dist)
    dist[:, 0] = np.arange(n1)
    dist[0, :] = np.arange(n2)

    for i in range(1, n1):
        for j in range(1, n2):
            dist[i, j], op[i, j] = min(
                (dist[i - 1, j] + 1, 1),
                (dist[i, j - 1] + 1, 2),
                (dist[i - 1, j - 1] + delta(word1[i - 1], word2[j - 1]), 3)
            )
    
    return dist, op
```

```python
def get_sequence(dist, op):
    result = []
    i, j = dist.shape
    i -= 1
    j -= 1
    while(i > 0 and j > 0):
        operation = op[i, j]
        result.append((i - 1, j - 1, operation))

        i -= operation & 1
        j -= operation >> 1 & 1
        operation = op[i, j]
    
    return reversed(result)
    
```

### Zadanie 2

```python
def distance(word1, word2):
    dist, op = alghorithm(word1, word2)
    
    print(word1)
    word = ""
    hint = ""
    for i, j, operation in get_sequence(dist, op):
        if operation == 3:
            if word1[i] == word2[j]:
                word += word1[i]
                continue
            current = word2[j]
            hint = word1[i] + " -> " + word2[j]
        elif operation == 1:
            current = ""
            hint = "   - " + word1[i]
        else:
            current = word2[j]
            hint = "   + " + word2[j]
        print(word + "{" + current + "}" + word1[i+1:] + "\t\t" + hint)
        word += current
    print(word2)
    return dist[-1, -1]
```

### Zadanie 4

```python
def lcs(seq1, seq2):
    sequence = []
    
    dist, op = alghorithm(seq1, seq2, lambda el1, el2: delta(el1, el2) * 2)
    for i, j, op in get_sequence(dist, op):
        if op == 3:
            sequence.append(seq1[i])
    
    return sequence
    
```

```python
# Nie jest to w 100% zgodne z opisem w treści zadania,
# w następnej sekcji można zobaczyć jak działa.

def diff(seq1, seq2):
    sequence = []
    
    dist, op = alghorithm(seq1, seq2, lambda el1, el2: delta(el1, el2) * 2)
    for i, j, op in get_sequence(dist, op):
        if op == 1:
            sequence.append(("<", seq1[i]))
        if op == 2:
            sequence.append((">", seq2[j]))
    
    return sequence
    
```
